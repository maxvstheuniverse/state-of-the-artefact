```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
data_path = "../data/output/output_2020-06-23T16-29-34.npy"

x = np.load(data_path, allow_pickle=True)

round_ids = np.array(x[:, 0])
z_means = np.array(x[:, 3])
```

```python
df = pd.DataFrame()
df.insert(0, 'round_id', round_ids)
df.insert(1, 'z_means', z_means)
df.insert(2, 'z', np.array(x[:, 5]))
```

```python
df
```

```python
m = df['z_means'].to_numpy()
m
```

```python
m.shape
```

```python
duplicate = set()
count = 0
for i, a in enumerate(m):
    for j, b in enumerate(m):
        if i <= j:
            count += 1
            continue
        
        if np.array_equiv(a, b):
            key = i
            if key not in duplicate:
                duplicate.add(i)

print(len(duplicate), count)
```
